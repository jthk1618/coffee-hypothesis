### Intro<a id='intro'></a>

I love black coffee. I've been through short phases of drinking other stuff but I always returned to filters, espresso, black coffee eventually and am now at a place where I can't imagine my tastes ever diverging again. I also like a level of white noise when I work, I like walking between periods of focus throughout the day and I dislike offices. Naturally, cafes have become my mobile workspaces where, over the years, it has come to feel almost as though I pay for desk rent and the coffee is a complimentary something on the side. For me, the cafe has become a symbolic trigger and pragmatic avenue for productivity more than all that is coffee and the surplus enjoyment of ritual is enjoyment. 

Another love of mine is feeling energised and alive every day - high quality and sufficient sleep. Perhaps due to the years of being fatigued and feeling zombie like in early high school for years on end, my sleep schedule has become one of my most preciously, jealously, zealously guarded possessions. Whatever the motivation, through years of experimentation I've come to understand my sleep cycle and variables that impact it quite deeply. In my endless pursuit for the best sleep I've had to compromise on social things, exercise things, surprising things and of course, coffee. Concomitant to actualising one is the imposition of a set of constraints on the other. Coffee is bounded in volume and timing of consumption, sleep is bounded in time, timing, quality and for both, compromise this way or that is always a little heartbreaking.

Currently, I have 2-3 coffees per day and typically have my last coffee before 11am. I believe this keeps my sleep quality high, but it means that I lose out on afternoon cafe opportunities. This isn't an enjoyment opportunity cost so much as it is a productivity one. Locking myself out of the cafe option in the afternoon can be constraining with regards to where I can be/go to get things done.

The nature of this analysis is probably becoming clearer by now, this is a matter of economising on a tradeoff with some nuances which I'll outline in a moment. I'll start by describing the nature of the problem in general terms, then get into the more personal, granular details.

### Guns and Butter <a id="guns-and-butter"></a>

At each level, there's a few things to consider. Some collection of these things are going to be opposing ends of the same spectrum, but that's hard to discern just now. Let's just list some sensible measures to begin.

Coffee measures:
- Coffee volume
- Coffee kind (espresso caffeine < v60)
- Latest coffee

Sleep measures:
- sleep score
- minutes of rem
- minutes of deep

Overarching measure:
- Productivity
- Subjective day to day feel
- Actual output

An example of what I mean when I say some of these things are opposing ends of the same spectrum would be if say, hypothetically, latest coffee and minutes of deep sleep are clearly negatively correlated. That's the kind of hypothesis that can be tested later on.

**High Level** <a id="high-level"><a/>

It's common knowledge that coffee has a half life of 6-8 hours. This is pretty constraining to start. I've heard Matthew Walker (sleep expert) mention that having your last cup of coffee 14 hours before bed time is ideal, which seems sensible given a 7 hour half life. At this level, later coffee would also mean worse sleep. On the other side of that tradeoff, the later coffee would mean a wider distribution of productivity hours throughout the day (6am - 4pm). I strongly believe that interweaving rest with work, ceteris paribus, is more effective in almost all pursuits for improvement. The analogy I like to use is that of eating: Eating enables potential nourishing but digestion is where the nourishing rubber hits the road. Gym monkeys are well aware of this dyad too. In the ballpark of the cellular level, training itself is destructive; only in recovery does the muscle itself gain access to greater tensility resources. Playing irresponsibly fast and loose with the concept, git works in an analogous way. In my experience, skills are no different. [GTG](https://mtntactical.com/knowledge/mini-study-grease-the-groove-beats-density-for-push-up-pull-up-improvement/) is evidence enough for me, I've found it to be very effective both in physical, mental and work skills. This might partially explain to the reader why I value the dispersed productivity hours over a day of density.

**Coffee Jouissance** <a id="coffee-jouissance"><a/>

It seems quite likely that [coffee preferences are associated with caffeine metabolism](https://www.sciencedirect.com/science/article/abs/pii/S0950329316301161). As it happens, all of my tastes in coffee would suggest that I am a fast metaboliser. If it is the case that I am a faster metaboliser, would it not make sense that I'd be able to have coffees later in the day without sacrificing too much / any sleep quality?

**Source Code** <a id="source-code"><a/>

As it turns out, I can check. I did a DNA test a while ago and I have some insight. I am an AA on the rs762551 for CYP1A2. Apparently this suggests that I am a fast caffeine metaboliser. It also turns out that having this particular star sign means that certain substances called inducers accelerates my caffeine metabolism. One such inducer is present in cruciferous vegetables which I happen to consume regularly. As a side note, it seems that the reason why cruciferous vegetables (similarly for mdma, diesel exhaust and hetrocyclic amines like tobacco smoke and charred meat) raise the CYP1A2 activity is becuase they're toxic. This lines up with what I already understand about [cruciferous veggies](https://www.cell.com/trends/neurosciences/comments/S0166-2236(06)00200-1). Caffeine also boosts atheletic performance in fit/strong men with the AA variation.

There's more DNA details I have access to, but let's stick to just this since the CYP1A2 gene is responsible for 90% of caffeine metabolism. The company that did my DNA test suggests that coffee within the time window of 6am - 6pm won't affect my sleep. The other interesting recommendation they make is that eating more cruciferous veggies as a means of speeding up metabolism could be useful. Unfortunately I don't have personal data on that.

**Conclusion**

All things considered, I have reason to believe that I might be losing out by bounding my coffee consumption hours to 11am and earlier. In the past, I have kept my coffee window open to as late as 2 pm which from memory was ok, but not ideal. If I was to return to the wider window, I would have more cafe time, more productive hours and perhaps feel better during the later morning. Furthermore, a deliberate and controlled use of cruciferous veggies may be strategically employed to extend my window.
